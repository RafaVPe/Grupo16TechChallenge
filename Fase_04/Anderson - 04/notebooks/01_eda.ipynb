## Metodologia diferente para o tratamento das variaveis e construção do modelo de ML 

Ao fazer algumas pequisas decidi adotar uma metodologia diferente e mais profissional nesta parte. 

A ideia é construir um Pipeline automatico que trasnforma as string em valores númericos de forma organizada sem a necessidade de um map manual como havia feito antes. 

```python
# Importações

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import StratifiedKFold, cross_val_score
import joblib
```

```python
df_obesity1 = pd.read_csv("../data/raw/Obesity.csv")
df_obesity1.head()
```

```python
cols_to_round = ['FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
df_obesity1[cols_to_round] = df_obesity1[cols_to_round].round()
```

```python
df_obesity1[cols_to_round] = df_obesity1[cols_to_round].astype(int)
```

```python
df_obesity1['FAF'].unique()
```

```python
# Preparando X e y sem tranformação manual

X = df_obesity1.drop("Obesity", axis=1)
y = df_obesity1["Obesity"]
```

```python
X.shape
```

```python
y.shape
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
#Divisão entre treino e teste 

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y  # O stratify=y é importante porque a target está bem distribuída.
)
```

### Criando o pipeline

```python
numeric_features = [
    "Age", "Height", "Weight", "FCVC", "NCP", "CH2O", "FAF", "TUE"    
]

categ_features = [ 
    "Gender" ,"family_history", "FAVC", "CAEC", "SMOKE", "SCC", "CALC", "MTRANS"
      ]

preprocessor = ColumnTransformer(
    transformers=[
        ("Num", StandardScaler(), numeric_features),
        ("Cat", OneHotEncoder(handle_unknown="ignore"), categ_features)
    ]
)
model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("classifier", RandomForestClassifier(random_state=42))
    ]
)
```

Utilizei o ColumnTransformer para identificar as colunas categóricas, nela o StandarScaler padroniza as colunas númericas (mesmo sem necessidade para modelos como o Random Forest), o OneHotEncoder transforma essas colunas em colunas binárias e o modelo vai receber tudo já convertido em números. 

Isso serve para evitar transformações manuais e permitir com que o Smtreamlit trabalhe mais facilmente. 

### Modelo: Random Forest Classifier

```python
model.fit(X_train, y_train)
```

#### Visualização dos dados tranformado

```python
# Pegando os nomes das novas colunas 
X_train_transformed = model.named_steps["preprocessor"].transform(X_train)

columns_names= model.named_steps['preprocessor'].get_feature_names_out()

X_train_df = pd.DataFrame(X_train_transformed, columns=columns_names)
X_train_df.head()
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

#### Classfication report e Matriz de confusão

```python
print(classification_report(y_test, y_pred))
```

```python
matriz_de_confusao = confusion_matrix(y_test, y_pred) 

plt.figure(figsize=(8,6))
sns.heatmap(matriz_de_confusao, annot= True, fmt='d', cmap = 'Blues')
plt.xlabel("Predito")
plt.ylabel("Real")
plt.title("Matriz de confusão")
plt.show()
```

#### Validação cruzada 

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(model, X, y_encoded, cv=skf, scoring="accuracy")

print("Acurácias por fold:", scores.round(2))
print("Média da acurácia:", scores.mean().round(2))
print("Desvio padrão:", scores.std().round(2))
```

O modelo foi validado utilizando a validação cruzada estratificada com 5 folds eliminando a sorte estatística de um único split.
Desvio padrão baixo indica que o modelo está estável e generaliza bem. 

#### Utilizando Feature Importance 

```python
feature_names = model.named_steps["preprocessor"].get_feature_names_out()
importances = model.named_steps["classifier"].feature_importances_

feat_imp = pd.DataFrame({
    "feature": feature_names,
    "importance": importances
}).sort_values(by="importance", ascending=False)

feat_imp.head(50)
```

A utilização de Feature Importance serve para medir o quanto cada variável ajudou o modelo a separar corretamente as classes, quanto maior, mais importante. 

### Modelo: Regressão logística

```python
from sklearn.linear_model import LogisticRegression

log_model = Pipeline(
    steps=[
       ("preprocessor", preprocessor),
        ("classifier", LogisticRegression(max_iter = 2000, random_state = 42))
    ]
)
```

```python
log_model.fit(X_train, y_train)
```

```python
y_pred_log = log_model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_log))
```

```python
mc_log = confusion_matrix(y_test, y_pred_log)

plt.figure(figsize=(8,6))
sns.heatmap(mc_log, annot= True, fmt='d', cmap = 'Blues')
plt.xlabel("Predito")
plt.ylabel("Real")
plt.title("Matriz de confusão- Regressão logística")
plt.show()
```

```python
scores_log = cross_val_score(log_model, X, y_encoded, cv=skf, scoring="accuracy")

print("Acurácias por fold:", scores_log.round(2))
print("Média da acurácia:", scores_log.mean().round(2))
print("Desvio padrão:", scores_log.std().round(2))
```

### Modelo: XGBoost

```python
from xgboost import XGBClassifier 

xgb_model = Pipeline(
    steps = [
        ("preprocessor", preprocessor),
        ("classifier", XGBClassifier(eval_metric="mlogloss", random_state=42))
    ]
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
```

```python
mc_xgb=confusion_matrix(y_test, y_pred_xgb)

plt.figure(figsize=(8,6))
sns.heatmap(mc_xgb, annot= True, fmt='d', cmap = 'Blues')
plt.xlabel("Predito")
plt.ylabel("Real")
plt.title("Matriz de confusão - XGBoost")
plt.show()
```
